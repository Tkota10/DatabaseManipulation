# Import Proper Libraries

```python
import pandas as pd
```

# Render the Data

```python
df = pd.read_csv("Final_Output.csv")
df
```

# Run Command To Find Duplicates

```python
Output = df.sort_values("distance",ascending=False).drop_duplicates('upload_ID', keep='last').sort_index()
Output.set_index("upload_ID", inplace=True)
Output.head(20)
```

# Write to a csv file

```python
Output.to_csv("Final_Output_Cleansed.csv")
```
